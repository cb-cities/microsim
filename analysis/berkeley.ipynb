```python
def write_config_file(path, config):
    file_path = path + 'command_line_options.ini'
    with open(file_path, 'w') as f:
        f.write('[General]\n')
    
    
    with open(file_path, 'a') as the_file:
        for k,v in config.items():
            line = k + '=' + v +'\n'
            the_file.write(line)
    print ('write to ' + file_path)
```
