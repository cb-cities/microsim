```python
import numpy as np
import pandas as pd
```

```python
berkeley_orig_od = pd.read_csv('day_time_od.csv')
berkeley_orig_od
```

```python
berkeley_processed_od = berkeley_orig_od[['origin_node_id', 'destin_node_id']].copy().rename(columns={
    'origin_node_id':'origin', 'destin_node_id':'destination'
})
berkeley_processed_od
```

```python
berkeley_processed_od['dep_time'] = np.random.choice(range(0, 3600, round(3600/len(berkeley_processed_od))), size=len(berkeley_processed_od))
berkeley_processed_od.to_csv('berkeley_processed_od.csv', index=False)
```
